## Funciones de Preprocesamiento (incluidas directamente en el Notebook)

Debido a restricciones de Kaggle para cargar módulos externos, la librería `libreria_Nelson_Casa.py` se incorpora directamente en esta celda.


```python
# =========================================
#  Librería de Preprocesamiento (Kaggle Safe)
# =========================================

import pandas as pd
import re
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

# Tokenizador sin modelo 'punkt'
tokenizer = RegexpTokenizer(r'\w+')

# Stopwords manuales para inglés (versión reducida)
stop_words_en = {
    "the","a","an","and","or","if","in","on","at","of","this","that","to",
    "is","are","was","were","be","been","it","for","with","as","by","from",
    "but","not","no","so","too","very","can","could","should","would","our",
    "your","their","my","his","her","its","them","they","he","she","we","you"
}

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocesamiento_en(texto):
    """Función de preprocesamiento compatible con Kaggle (sin descargas NLTK)."""

    if pd.isna(texto):
        return ""

    # eliminar HTML
    texto = re.sub(r'<[^>]+>', '', texto)

    # solo letras
    texto = re.sub(r'[^a-zA-Z\s]', ' ', texto)

    # normalizar espacios
    texto = re.sub(r'\s+', ' ', texto).strip()

    texto = texto.lower()

    # tokenización sin punkt
    tokens = tokenizer.tokenize(texto)

    # stopwords
    tokens = [word for word in tokens if word not in stop_words_en and len(word) > 1]

    # lematización
    tokens_lemma = [lemmatizer.lemmatize(word) for word in tokens]

    # stemming
    tokens_stem = [stemmer.stem(word) for word in tokens_lemma]

    return " ".join(tokens_stem)
```

## 3. Preprocesamiento de textos

En este dataset:
- `movies` contiene reseñas de críticos (`review_content`)
- `reviews` contiene la información principal de las películas

Procedimiento:
1. Agrupamos todas las reseñas por película  
2. Las unimos a la tabla principal  
3. Construimos un texto combinado por película  
4. Aplicamos el preprocesamiento  


```python
# Asegurar que no hayan nulos
peliculas["movie_title"] = peliculas["movie_title"].fillna("").astype(str)
peliculas["genres"] = peliculas["genres"].fillna("").astype(str)
peliculas["movie_info"] = peliculas["movie_info"].fillna("").astype(str)
peliculas["critics_consensus"] = peliculas["critics_consensus"].fillna("").astype(str)

# Construcción correcta del texto para RI
peliculas["combined_text"] = (
    peliculas["movie_title"] + " " +
    peliculas["genres"] + " " +
    peliculas["movie_info"] + " " +
    peliculas["critics_consensus"]
)

# Preprocesar
peliculas["processed_text"] = peliculas["combined_text"].apply(preprocesamiento_en)

peliculas[["movie_title", "processed_text"]].head()
```

## 4. Construcción del Sistema de Recuperación
Utilizamos:

- **TF–IDF** para representar los textos
- **Cosine Similarity** para medir relevancia

Este enfoque es clásico y suficiente para los requisitos del examen.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(peliculas["processed_text"])
```

## 5. Función de Búsqueda
Toma una consulta, la procesa y devuelve las películas más relevantes.


```python
def buscar(query, top=10):
    query_clean = preprocesamiento_en(query)
    query_vec = vectorizer.transform([query_clean])

    scores = cosine_similarity(query_vec, tfidf_matrix)[0]
    
    peliculas["score"] = scores
    return peliculas.sort_values("score", ascending=False)[["movie_title", "score"]].head(top)
```

## 6. Simulación de Consultas

Probamos el sistema con las consultas del examen:
- Películas sobre viajes espaciales
- Películas para ver en familia

```python
print("Search: Movies about space travel")
display(buscar("space travel movies"))

print("\nConsult: Movies to watch as a family")
display(buscar("family movies"))
```

Esto significa que el modelo funciona especialmente bien en búsquedas directas, donde la relación entre las palabras de la consulta y el contenido del dataset es evidente.

Esto demuestra que para consultas más abstractas o temáticas, el modelo tiene dificultades para capturar la intención real del usuario, ya que depende completamente de coincidencias literales de palabras.
