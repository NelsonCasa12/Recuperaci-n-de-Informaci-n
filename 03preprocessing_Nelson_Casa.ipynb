#### 1. Limpieza

Limpiar los documentos de caracteres que no corresponden

```python
doc = df.iloc[0]['review']
doc
```

```python
doc.replace('<br />', '').replace('>', '').replace('/', '')
```

```python
import re
```

```python
def clean_text(doc):
    return re.sub(pattern=r'<.*?>', repl='', string=doc).replace('.', ' ').replace(',', '').replace('(', '').replace(')', '').replace('"', '').replace("'", '').replace("\x08", '')
```

```python
df['review'].apply(clean_text)
```

```python
doc.split()
```

#### 2. Normalización

Convertir todos los tokens a minúsculas.

Elimina puntuación y símbolos no alfabéticos.

En esta etapa transformamos el texto preprocesado hacia una forma más homogénea y consistente.
El objetivo es reducir la variabilidad lingüística y dejar únicamente características esenciales

```python
import re
import string
```

```python
def normalize_text(doc):
    doc = re.sub(r'[^a-z\s]', ' ', doc.lower())
    return re.sub(r'\s+', ' ', doc).strip()
```

Aplicar normalización al documento de ejemplo

```python
cleaned_doc = clean_text(doc)
normalized_doc = normalize_text(cleaned_doc)

print(normalized_doc)
```

#### 3. Eliminación de Stopwords

Eliminar las palabras vacías (stopwords) usando una lista estándar de la librería _nltk_.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
# Tokenizar
tokens = normalized_doc.split()
# crear una nueva lista solo con las palabras que NO están en stop_words
tokens_sin_stopwords = [palabra for palabra in tokens if palabra not in stop_words]
# Imprimir cuántas palabras se eliminaron
print(f"Palabras originales: {len(tokens)}")
print(f"Palabras sin stopwords: {len(tokens_sin_stopwords)}")
print("\n--- Ejemplo de tokens sin stopwords ---")
print(tokens_sin_stopwords)
```

#### 4. Stemming

Reducir el espacio de palabras

```python
from nltk.stem import porter

stemmer = porter.PorterStemmer()
```

```python
tokens_stemmed = [stemmer.stem(palabra) for palabra in tokens_sin_stopwords]

print("\n--- Ejemplo de tokens con stemming ---")
print(tokens_stemmed)
```

```python
texto_final_stemmed = ' '.join(tokens_stemmed)

print(texto_final_stemmed)
```
